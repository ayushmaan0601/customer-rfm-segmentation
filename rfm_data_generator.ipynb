```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

random.seed(42)
np.random.seed(42)

# Settings
n_customers = 1000
n_transactions = 8000
reference_date = datetime(2024, 12, 31)

# Generate customers
customer_ids = [f"CUST{str(i).zfill(4)}" for i in range(1, n_customers+1)]

# Generate transactions
data = []
for _ in range(n_transactions):
    customer_id = random.choice(customer_ids)
    days_ago = random.randint(1, 365)
    purchase_date = reference_date - timedelta(days=days_ago)
    amount = round(random.uniform(10, 500), 2)
    data.append([customer_id, purchase_date.strftime('%Y-%m-%d'), amount])

df = pd.DataFrame(data, columns=['customer_id', 'purchase_date', 'amount'])
df.to_csv('transactions.csv', index=False)
print(f"Generated {len(df)} transactions for {n_customers} customers")
print(df.head())
```
